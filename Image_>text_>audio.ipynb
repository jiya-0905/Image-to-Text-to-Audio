```python
import pytesseract
from PIL import Image #image processing
import gtts #google text to speech
```

```python
image = Image.open('/content/LOC.png')
print(image)
```

```python
content = pytesseract.image_to_string(image)
print(content) #shows minor errors due to lack of clarity in image as i have used a screenshot
audio= gtts.gTTS(content)
audio.save("LOC.mp3")
```
